## 4. Dataset, Splitting the dataset
The dataset was split once into training, validation, and testing sets using a 60/20/20 ratio. The split was saved to three shared CSV files: train_data.csv, val_data.csv, and test_data.csv. This ensured that all group members used the exact same data in their experiments and that the results were comparable.

The task is a multi-class classification problem. The target variable contains several attack categories, such as Normal Traffic, DDoS, DoS, Port Scanning, Brute Force, Web Attacks, and Bots. The labels were encoded into numerical values before training the models.

## 7. SMOTE
SMOTE was tested as an optional preprocessing step for kNN. The model was trained both with SMOTE (applied only to training data) and without it, and both options were evaluated on validation accuracy. The findings showed whether synthetic oversampling improved or degraded the model's ability to detect minority attack classes. The best configuration (with or without SMOTE) was selected based on validation accuracy and then tested on the test set.


---

## 8. Scaling, Polyfeatures, PCA, LDA, Kernel PCA


Scaling: StandardScaler and MinMaxScaler were tested before training kNN, as distance-based models are sensitive to feature magnitude.

Polynomial Features: Not tested for kNN, as polynomial feature expansion is not critical for distance-based models.

Dimensionality Reduction: PCA, LDA, and Kernel PCA were not tested. The dataset already uses 10 carefully selected features, so additional dimensionality reduction was not necessary.


---

## 9. Training setups
The kNN model was trained by testing different combinations of scalers, SMOTE options, and kNN hyperparameters. The best configuration was selected based on validation accuracy. The best model configuration was then retrained and evaluated on the test set using the same scaler and SMOTE option that produced the highest validation accuracy.
