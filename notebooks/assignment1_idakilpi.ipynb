### 7.Answer

SMOTE was tested as an optional pipeline step for both Decision Tree and Random Forest. It was used only on the training part of the data and compared against `passthrough`, meaning no SMOTE. The purpose was to reduce the effect of class imbalance, because some attack types had much fewer samples than `Normal Traffic`.

The overall finding was that SMOTE was worth testing, but it was not necessarily the best choice for the final tree-based models. Decision Tree and Random Forest already performed well without synthetic oversampling, especially when class weighting was also tested. SMOTE may help minority-class recall, but it can also hurt precision or accuracy, so the final choice was based on validation accuracy.
