## 7. SMOTE

SMOTE was tested as an optional preprocessing step in the model pipelines. It was applied only to the training data, not to the validation or test data. In the pipelines, SMOTE was compared against passthrough, which means that SMOTE was not used. The reason for testing SMOTE was the strong class imbalance in the dataset: Normal Traffic had more samples than rare attack classes such as Bots and Web Attacks.

SMOTE was useful to test, but it was not selected as the best option for most final models. For Logistic Regression, SVM, kNN, Naive Bayes, Decision Tree and Random Forest, the best validation accuracy was generally achieved without SMOTE. Tree-based models already performed well without synthetic oversampling, and Logistic Regression and SVM did not benefit enough from SMOTE in terms of overall accuracy. In some cases, SMOTE improved recall for minority classes, but it also reduced precision or accuracy. Therefore, the final choice was made based on validation performance, while also considering recall and F1-score because of the imbalanced class distribution.
